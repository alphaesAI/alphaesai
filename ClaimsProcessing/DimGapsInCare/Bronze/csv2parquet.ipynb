```python
# Databricks Notebook Source
# Setup widgets to accept values passed from the orchestrator
dbutils.widgets.text("csv_path", "", "Input CSV Path")
dbutils.widgets.text("catalog_path", "", "Target Catalog Path")

# Retrieve the values dynamically
csv_path = dbutils.widgets.get("csv_path")
catalog_path = dbutils.widgets.get("catalog_path")
```

```python
# Read CSV
df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(csv_path)

# Write as parquet table
df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable(f"{catalog_path}")

print(f"Written {df.count()} rows to {catalog_path}")
```
