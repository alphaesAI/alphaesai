```python
csv_path = "/Workspace/Repos/logi@openhealthagents.org/alphaesai/ClaimsProcessing/source/DimQualityEvent/pending/qualityevent.csv"
catalog_path = "claimsprocessing.bronze.quality_measures"
```

```python
# Read CSV
df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(csv_path)

# Write as parquet table
df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable(f"{catalog_path}")

print(f"Written {df.count()} rows to {catalog_path}")
```
