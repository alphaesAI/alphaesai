```python
from pyspark.sql import DataFrame
```

```python
def output_validation_error(df_error: DataFrame, error_path: str) -> None:
    df_error.coalesce(1).write.mode("append").option("delimiter", ",").option("header", "true").csv(error_path)

def cosolidate_error(df_error: DataFrame, error_path: str) -> None:
    output_validation_error(df_error, error_path)
```
