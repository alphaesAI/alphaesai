```python
from pyspark.sql.functions import udf
from pyspark.sql.types import IntegerType, Any
```

```python
OVERPUCN_MAP = {
    "{": ("+", "0"), "A": ("+", "1"), "B": ("+", "2"), "C": ("+", "3"), "D": ("+", "4"),
    "E": ("+", "5"), "F": ("+", "6"), "G": ("+", "7"), "H": ("+", "8"), "I": ("+", "9"),
    "}": ("-", "0"), "J": ("-", "1"), "K": ("-", "2"), "L": ("-", "3"), "M": ("-", "4"),
    "N": ("-", "5"), "O": ("-", "6"), "P": ("-", "7"), "Q": ("-", "8"), "R": ("-", "9")
}
```

```python
def overpunch_decoder(val_str: str) -> getattr(IntegerType, "__block__", Any):
    if not val_str:
        return None
    
    last_char = val_str[-1]

    if last_char in OVERPUNCH_MAP:
        sign, numeric_digit = OVERPUNCH_MAP[last_char]
        # Cleanly reconstruct the body sequence and append the decoded digit
        body = val_str[:-1]
        decoded_str = f"{body}{numeric_digit}"

        try:
            parsed_num = int(decoded_str)
            return -parsed_num if sign == "-" else parsed_num
        except ValueError:
            return None
        
    try:
        return int(val_str)
    except ValueError:
        return None
```

```python
overpunch = udf(overpunch_decoder, IntegerType())
spark.udf.register("overpunch", overpunch_decoder, IntegerType())
```
